# 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#1
```

```python
%config InlineBackend.figure_formats = {'png', 'retina'}  # for high-reso graph
plt.rcParams['font.family'] = 'Yu Mincho' # for Japanese in graph (Win)

csv_in = 'data/covid-19-pref-area-20200403.csv'
```

# 2

```python
df = pd.read_csv(csv_in, sep=',', skiprows=7, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

# 3

```python
ser_aclose = df['PCR']
plt.hist(ser_aclose, bins=10)
plt.xlabel('PCR')
plt.ylabel('Number of Prefs')
plt.show()
```

# 4

```python
df2 = df[['Infected','Discharged']]
plt.boxplot(df2, labels=df2.columns)
plt.title('Number of people')
plt.show()
```

# 5

```python
ser_x = df['Infected']
ser_y = df['Discharged']
plt.scatter(ser_x, ser_y, marker='o')
plt.xlabel('Infected')
plt.ylabel('Discharged')
plt.show()
```

# 6

```python
df_area=df.groupby('Area').sum()
display(df_area)
plt.pie(df_area['Infected'],autopct="%.1f%%", labels=df_area.index)
plt.show()
```

# 7

```python
df_area['Infected_rate'] = df_area['Infected'] / df_area['Pop(100k)']

plt.bar(df_area['Infected_rate'].index, df_area['Infected_rate'], label='Average')
plt.ylabel('Infected per Pop(100k)')
plt.xticks(rotation=90)
plt.show()
```

# 8

```python
csv_in2 = 'data/covid-19-japan-20200407.csv'
df_jpn = pd.read_csv(csv_in2)
print(df_jpn.shape)
print(df_jpn.info())
display(df_jpn.head())
```

# 9

```python
df_jpn['Date'] = pd.to_datetime(df_jpn['Date'])
print(df_jpn.dtypes)
```

# 10

```python
plt.bar(df_jpn['Date'], df_jpn['Infected_day'], label='Average')
plt.ylabel('Infected per day')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
ser_7d_ave = df_jpn['Infected_day'].rolling(window=7).mean()

plt.bar(df_jpn['Date'], ser_7d_ave, label='Average')
plt.ylabel('Infected per day(ave7)')
plt.xticks(rotation=90)
plt.show()
```
