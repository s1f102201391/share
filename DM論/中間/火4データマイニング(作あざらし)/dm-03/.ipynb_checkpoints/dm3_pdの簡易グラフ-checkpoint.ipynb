## Create simple plot with Pandas  

### Import libraries

```python
import pandas as pd

#pandasしか読み込んでいないことに注意
```

### Parameters

```python
csv_in1 = 'data/2045_2017-utf8.csv'
```

### Read CSV file

```python
df1 = pd.read_csv(csv_in1, sep=',', skiprows=9, header=0)
df1['Date']=pd.to_datetime(df1['Date'], format="%Y-%m-%d")
print(df1.shape)
print(df1.info())
display(df1.head())
```

### Add 25days moving average line  

```python
df1['25d_mov_ave'] = df1['AdjClose'].rolling(window=25).mean()
display(df1.head(30))

#25日移動平均線の列の追加
#0~24日目までの25日分の平均が24列目なので、0~23は値がない
```

### Line plot  

```python
df1.plot.line('Date', ['AdjClose', '25d_mov_ave'])
```

### Scatter plot  

```python
df1.plot.scatter('High', 'Low')
```
