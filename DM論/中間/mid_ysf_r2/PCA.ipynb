```python
#Cell_1.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
plt.rcParams['font.family'] = 'Meiryo'
#plt.rcParams['font.family'] = 'Hiragino Sans' # for macOS
```

```python
#Cell_2.
csv_in = '2016-2017_NBA_stats_data.csv'

# To show all rows and columns in the results 
pd.options.display.max_columns=999
pd.options.display.max_rows=999
```

```python
#Cell_3.
df = pd.read_csv(csv_in, sep=',', skiprows=27, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
#Cell_4.
df['over_half']=0
for i in range(df.shape[0]):
    if(df.iloc[i,2] > df.iloc[i,3]):
        df.loc[i,'over_half'] =1

display(df.head())
```

```python
#Cell_5.
print(df.columns)
```

```python
#Cell_6.
dfX = df.loc[:, 'GP':' +/-']
display(dfX.head())
```

```python
#Cell_7.
sc = StandardScaler()
X_std = sc.fit_transform(dfX)
```

```python
#Cell_8.
n_pca = 7
pca = PCA(n_components=n_pca)
X_pca = pca.fit_transform(X_std)
```
