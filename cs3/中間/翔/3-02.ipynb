## 行列全体の取り出し

・複数列の取り出し(iloc)

```python
import numpy as np
import pandas as pd
```

```python
csv_in = 'wholesale_customers.csv'
df = pd.read_csv(csv_in, sep=',', skiprows=13, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
print(df.index)
```

```python
print(df.columns)
```

```python
ser_region = df['Region']
print(type(ser_region))
print(ser_region)
```

```python
df2 = df[ ['Region', 'Milk', 'Grocery'] ]
print(type(df2))
display(df2.head())
```

```python
df3 = df.loc[ 1:3, 'Milk':'Frozen' ]
print(type(df3))
display(df3)
```

```python
display(df.head(3))
df4 = df.loc[ :, 'Milk': ]
display(df4)
df5 = df.loc[ :, :'Frozen' ]
display(df5)
```

```python
display(df.head(3))
```

```python
df6 = df.iloc[ 1:4, 3:6 ]
#df6 = df.iloc[ 1:4, -5:-2 ]
display(df6)
df6 = df.iloc[ -439:-436, -5:-2 ]
display(df6)
```

```python
display(df.iloc[ -440:-435:2, 5:2:-1 ])
```
