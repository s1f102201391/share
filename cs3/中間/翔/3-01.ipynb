## DataFrame のつくりかた

```python
df = pd.DataFrame([ [0, 1, 2], [3, 4, 5] ], columns = ['c0', 'c1', 'c2'])
print(type(df)) 
print(df.shape) 
display(df)
```

```python
df2 = pd.DataFrame( [ [0, 1, 2], [3, 4, 5] ], columns = ['c0', 'c1', 'c2'],  index= [ 'ro', 'r1'])
display(df2)
```
