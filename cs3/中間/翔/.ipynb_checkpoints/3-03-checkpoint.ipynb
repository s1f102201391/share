## データの様子の把握・可視化
・基本統計量

・数値変数、カテゴリー変数(describe)

・データ型の変更

・箱ひげ図

・ヒストグラム

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_in = '../cs3-02/wholesale_customers.csv'
df = pd.read_csv(csv_in, sep=',', skiprows=13, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
df_num = df.loc[:, 'Fresh':]
display(df_num.head())
```

```python
print(np.sum(df_num, axis=0)) #index軸に対する計算(各列に対する和)
```

```python
print(np.sum(df_num, axis=1)) #columns軸に対する計算(各行に対する和)
```

```python
ave_milk = df_num['Milk'].mean()
print(ave_milk)
ave_milk = np.mean(df_num['Milk'])
print(ave_milk)
```

```python
display(df.describe())
```

```python
display(df.describe(exclude=[np.number]))
```

```python
df['Channel'] = df['Channel'].astype('category')
print(df.info())
```

```python
display(df.describe())
display(df.describe(exclude=[np.number]))
```

# 箱ひげ図

```python
plt.boxplot(df_num, labels=df_num.columns)
plt.xticks(rotation=90)
plt.ylabel('Annual spending [m.u.]')
plt.show()
```

# ヒストグラム

```python
plt.hist(df_num['Fresh'])
plt.xlabel('Annual spending [m.u.]')
plt.ylabel('Frequency')
plt.show()
```
