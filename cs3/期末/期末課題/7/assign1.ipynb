```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_in = 'boston_small.csv'
df = pd.read_csv(csv_in, sep=',', skiprows=8, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
df_corr = df.corr()
display(df_corr)
```

```python
print(df_corr.at['CRIM', 'INDUS'])
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_corr, annot=True, fmt='.2f', cmap='bwr', vmin=-1.0, vmax=1.0)
plt.show()
```

```python
df.corrwith(df['MEDV']).sort_values(ascending=False)
```

```python
plt.scatter(df['MEDV'], df['RM'])
plt.title('corr: {:.2f}'.format(df_corr.at['MEDV', 'RM']))
plt.xlabel('MEDV')
plt.ylabel('RM')
plt.show()
```
