```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from joblib import dump
```

```python
csv_in = 'iris.csv'
df = pd.read_csv(csv_in, sep=',', skiprows=2, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
X = df.drop(columns='class')  # features
y= df['class']  # true labels
print('X:', X.shape)
display(X.head())
print('y:', y.shape)
print(y.head())
```

```python
print(y.value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=16, stratify=y)
```

```python
print(X_train.shape)
print(y_train.shape)
print(y_train.value_counts())
```

```python
print(X_test.shape)
print(y_test.shape)
print(y_test.value_counts())
```

```python
rfc = RandomForestClassifier(n_estimators=10, max_depth=None,
                             random_state=16)     
```

```python
%%time
rfc.fit(X_train, y_train)
```

```python
ser_fi = pd.Series(rfc.feature_importances_, index=X_train.columns)
print(ser_fi.sort_values(ascending=False))
```

```python
plt.bar(X_train.columns, rfc.feature_importances_)
plt.ylabel('Importance')
plt.xticks(rotation=90)
plt.show()
```

```python
y_train_pred = rfc.predict(X_train)
display(pd.crosstab(y_train, y_train_pred))
```

```python
acc = accuracy_score(y_train, y_train_pred)
print('Accuracy for train data:', acc)
```

```python
y_test_pred = rfc.predict(X_test)
display(pd.crosstab(y_test, y_test_pred))
```

```python
acc = accuracy_score(y_test, y_test_pred)
print('Accuracy for test data:', acc)
```

```python
tag = 'iris'
model_file = 'rfc_{}.joblib'.format(tag)
dump(rfc, model_file)
```
