# RFR / ディープラーニングとCNN

##### RFRのプログラミング
##### Random Forest Regressor (RFR) + 発展
##### RFR (read model from file)
##### ディープラーニング

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from joblib import dump
```

```python
csv_in = 'student-por-gp.csv'
df = pd.read_csv(csv_in, sep=',', skiprows=0, header=0)
print(df.shape)
print(df.info())
display(df.head())
```

```python
X = df.drop(columns='G3')  # features
y = df['G3']  # true values    
print('X:', X.shape)
display(X.head())
print('y:', y.shape)
print(y.head())
```

```python
X['studytime'] = X['studytime'].replace(
    {'<2 hours':1, '2 to 5 hours':2,
     '5 to 10 hours':3, '>10 hours':4,
    }
)
display(X.head())
```

```python
X_dumm = pd.get_dummies(X, drop_first=True)
print('X_dumm:', X_dumm.shape)
display(X_dumm.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_dumm, y, test_size=0.3, random_state=13)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
rfr=RandomForestRegressor(n_estimators=500, max_depth=None,
                          random_state=17)    
```

```python
%%time
rfr.fit(X_train, y_train)
```

```python
ser_fi = pd.Series(rfr.feature_importances_, index=X_train.columns)
print(ser_fi.sort_values(ascending=False))
```

```python
plt.bar(X_train.columns, rfr.feature_importances_)
plt.ylabel('Importance')
plt.xticks(rotation=90)
plt.show()
```

# 予測 発展

```python
y_train_pred = rfr.predict(X_train)
```

```python
y_min = np.append(y_train_pred, y_train).min()
y_max = np.append(y_train_pred, y_train).max()
y_margin = (y_max - y_min) * 0.1
y_min -= y_margin
y_max += y_margin
```

```python
plt.scatter(y_train_pred, y_train, alpha=0.3)
plt.plot([y_min,y_max], [y_min,y_max])
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(y_min, y_max)
plt.ylim(y_min, y_max)
plt.title('Train data')
plt.xlabel('Pred')
plt.ylabel('True')
plt.show()
```

```python
mse = mean_squared_error(y_train, y_train_pred)
print('MSE, RMSE for train data:', mse, np.sqrt(mse))
```

```python
y_test_pred = rfr.predict(X_test)
```

```python
y_min = np.append(y_test_pred, y_test).min()
y_max = np.append(y_test_pred, y_test).max()
y_margin = (y_max - y_min) * 0.1
y_min -= y_margin
y_max += y_margin
```

```python
plt.scatter(y_test_pred, y_test, alpha=0.3)
plt.plot([y_min,y_max], [y_min,y_max])
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(y_min, y_max)
plt.ylim(y_min, y_max)
plt.title('Test data')
plt.xlabel('Pred')
plt.ylabel('True')
plt.show()
```

```python
mse = mean_squared_error(y_test, y_test_pred)
print('MSE, RMSE for test data:', mse, np.sqrt(mse))
```

```python
tag = 'student-por'
model_file = 'rfr_{}.joblib'.format(tag)
dump(rfr, model_file)
```
