# (1)

```python
def caesar_cipher(text, shift):
    result = ""
    for char in text:
        if char.isalpha(): 
            if char.islower():
                result += chr((ord(char) - ord('a') + shift) % 26 + ord('a'))
            elif char.isupper():
                result += chr((ord(char) - ord('A') + shift) % 26 + ord('A'))
        else:
            result += char
    return result

plaintext = "INIAD"
shift_amount = 3
ciphertext = caesar_cipher(plaintext, shift_amount)

print(f": {plaintext}")
print(f": {ciphertext}")
```

# (2)

```python
def caesar_decipher(ciphertext, shift):
    return caesar_cipher(ciphertext, -shift)

ciphertext = "LQLDG"
shift_amount = 3
deciphertext = caesar_decipher(ciphertext, shift_amount)

print(f": {ciphertext}")
print(f": {deciphertext}")
```
