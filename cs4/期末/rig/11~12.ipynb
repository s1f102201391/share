# Lecture11 セキュリティの基本

### 情報セキュリティの概念
- 脅威：情報システムのセキュリティを脅かす原因
- インシデント；事件事故などトラブルが生じた場合
- 脆弱性：情報システムに存在する弱点
- リスク：何らかの事態が起こることに関する不確実性
### 情報の流出
- クラッキング：不正コンピュータ侵入しデータ破壊など
- スパイウェア：気づかれないうちに個人情報など第三者に送信するマルウェア
- ソーシャルエンジニアリング：技術ではなく、会話やメモを盗み聞き
- cookie悪用
### サイバー攻撃
- ウェブサイトの改ざん
- Dos、DDoS攻撃、大量の接続要求(SYN flood、F5 attack)



## CIA
- 機密性：アクセス権限がある人のみしか見れない状態を確保(/秘密が守られない)
- 完全性：情報破壊、改ざんなどされていない状態を確保(/メッセージが書き換えられる)
- 可用性：アクセス権限があるものが必要時に中断することなくアクセスできる状態であること(/メッセージが捨てられてしまう)


## セキュリティ攻撃(Attacks)
- 機密性を脅かす攻撃
    - スヌーピング
    - トラフィック分析
- 完全性を脅かす攻撃
    - Modification(変更)
    - マスカレード、なりすまし
    - Replaying(再生)
    - Repudiation(拒否)
- 可用性を脅かす攻撃
    - DoS(Denial of service)
- パッシブ攻撃、アクティブ攻撃
    - パッシブ：通信を傍受するだけの攻撃
    - アクティブ：何かデータを送信することで被害を引き起こす攻撃

## 暗号技術の基礎(p.52～)
- 暗号化：暗号という処理
- 暗号文：暗号化されたデータ
- 平文：暗号化される前のデータ
- 復文：暗号化から平文への変換
## 暗号化アルゴリズムと鍵
- 暗号化アルゴリズム：暗号化の手順、アルゴリズム
- 暗号化鍵：平文から暗号文へ変換する際に使う特殊な情報
- 復号鍵：復号する際の特殊な情報

# Lecture12 暗号化方式
## シーザー暗号
- アルファベットの各文字をkey文字ずつずつ後ろへずらす
- 平文a、暗号文c
- Additive Cipherと呼ばれる
- コードある
- (何文字目＋key)%26(アルファベット総数)
## ヴィジュネル暗号(p18)
- 多表式換字暗号：変換規則を切り替えながら変換
- 上段が平文、左段が鍵、交わる文字が暗号文
- 鍵は最後の文字を使ってもそのままループして使用
- 復号：左段の鍵と交わる暗号文をもとに上段の平文を求める
## 暗号解読(p.24)
- 総当たり攻撃、ブルートフォース攻撃
- 統計的攻撃
### バーナム暗号(p.31)
- 乱数列を暗号化鍵として利用
- 暗号化するつど乱数列を用意して暗号化カギとし、それと平文をビットごとに演算して暗号文を得る
- 長所：理論的には解読できない
- 短所：暗号化カギの配送の問題、乱数列を再利用できない、厳密な意味で乱数列を生成できるのか
### 排他的論理和
- ２つの入力のどちらか片方が真でもう片方が偽の時には結果が真となり、両方とも真あるいは両方とも偽の時は偽となる演算
## 共通鍵暗号化方式と公開鍵暗号化方式(p.37)
- 共通鍵暗号化方式、対称暗号方式、共通鍵暗号化方式、秘密鍵暗号方式：暗号化鍵と復号鍵が同一である暗号化方式
- 短所：あらかじめ通信する相手と鍵を安全に共有
- →公開鍵暗号化方式：暗号化に使う鍵と復号に使う鍵を別にする
    - 非対称暗号方式


## 共通鍵暗号化技術(p.48)
- ストリーム暗号化：疑似乱数生成器が生成する乱数を使い暗号化、バイト単位
- データ攪拌部と鍵生成部から構成され暗号をブロック単位で行う
- DES：64ビットのブロック単位で暗号化
- ３DES；DESを３回する
- AES:ブロック暗号化方式、128ビット
- RC4:ストリーム暗号化方式、可変長鍵
- ブロックチェイニング：個々のブロックに対して個別に暗号化を行う(CBC,OFB,CTR)
