### イーサネット
- 優先ネットワーク技術
- LANのデファクトスタンダード

- ふたつのイーサネット(イーサネット：共有メディア、スイッチングイーサネット：専用メディア)
#### CSMA/CDでのブロードキャストによるメディアアクセス
- CSMA；送信する前に聞く(チャンネルが静かか確かめる)、フレーム全体を送信
- CD：衝突が検出された場合に送信を終了する
- 指数バックオフ：バックオフ期間に送信者が同じフレームの迷信試行でフレーム衝突を繰り返すと指数関数的に増加する
- バイナリ指数バックオフ：最大25回

#### スイッチングイーサネット
- 全二重、半二重イーサネット
- バッファリング：過剰なフレームの一時格納
- MACアドレスフィルタリング：フレーム送信判断
- →自己学習、フォワーディング

## BLE(Bluetooth Low Energy)
- 低電力無線通信用の新しい技術
- 通信範囲100m、データレート1Mdps、ピーク電15mA(Bt時30mA)
## NFC(Near Field Communication)
- ふたつの電子デバイスが近接距離で通信を確立することを可能にする通信プロトコル
- 通信範囲10cm、データレート424kdps
## LPWA(Low-Power Wide Area)
- 省電力広域無線技術
- 通信範囲10km、データレート数10kbps以下、バッテリ寿命10年以上
